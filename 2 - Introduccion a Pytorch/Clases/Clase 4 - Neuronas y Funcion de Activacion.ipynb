```python
# Transformar una imagen - CREAR LA FUNCION
preprocess = transforms.Compose([
    transforms.Resize(128),
    transforms.CenterCrop(128),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])
# Es mejor trabajar con datos pequeños, 128 ~ 255
# Alinear al centro por resolucion
# Representar la imagen a tensor, para que sean numeros
# Normalizacion
    # Los datos se entregaron en el ejercicio, despues se calcularan

```

```python
img_t = preprocess(img)   # Aplicacion del metodo
batch = torch.unsqueeze(img_t, 0)
plt.imshow(img_t.permute(1, 2, 0))  # La permutacion es para cambiar la dimension
```

```python
inception.eval()
```

```python
out = inception(batch)
# out

# Como son datos abstractos, se puede ver que los numeros no hacen sentido a simple vista, pero para el modelo tienen sentido

# Aqui se genera un tensor de una dimension con muchos datos (1000 aprox) con x valor en peso
```

```python
# out
```
